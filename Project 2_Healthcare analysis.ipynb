# Data Cleaning

```python
data = raw_data.copy() # Making a copy of the original data
data
```

```python
data.shape # Number of rows and columns in the dataset
```

```python
data.columns # List of columns in the dataset
```

```python
data.isnull().sum() # Checking for Null values in the dataset
```

```python
data.duplicated().any() # Checking if there is any repeated data in the dataset
```

```python
data.dtypes # Data types of each columns of the dataset
```

```python
# Changing the data types of the columns to their appropriate one

data['Name'] = data['Name'].astype('string')
data['Gender'] = data['Gender'].astype('category')
data['Blood Type'] = data['Blood Type'].astype('string')
data['Medical Condition'] = data['Medical Condition'].astype('string')
data['Date of Admission'] = pd.to_datetime(data['Date of Admission'])
data['Doctor'] = data['Doctor'].astype('string')
data['Hospital'] = data['Hospital'].astype('string')
data['Insurance Provider'] = data['Insurance Provider'].astype('string')
data['Admission Type'] = data['Admission Type'].astype('category')
data['Discharge Date'] = pd.to_datetime(data['Discharge Date'])
data['Medication'] = data['Medication'].astype('string')
data['Test Results'] = data['Test Results'].astype('category')
```

```python
data.dtypes # new data types of the dataset
```

```python
data = data.rename(columns={'Billing Amount': 'Billing Amount($)'}) # changing the name of the Billing amount column
data
```

```python
data.info() # Summary of the datas in the dataset
```

```python
data.describe() # Statistical summary of the dataset
```

# Exploratory Analysis and Visualiztion

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (14,5)
```

```python
data['Year of Admission'] = data['Date of Admission'].dt.year # Creating a new column for the Year of admission alone
data.sample(2)
```

```python
sns.countplot(data=data, x='Year of Admission', hue='Gender')
plt.legend(bbox_to_anchor =(1.15, 1))
plt.ylabel('Patient count')
plt.title('Number of Patients admitted per year')
plt.grid();
```

```python
sns.scatterplot(data=data[data['Year of Admission']==2023], x='Age', y='Room Number', hue='Medical Condition')
plt.legend(title='Medical condition', bbox_to_anchor =(1.25, 1))
plt.title('2023 Relationship of age of admitted Patients with Room number and Medical condition');
```

```python
sns.countplot(data=data[data['Year of Admission']==2023], x='Blood Type', hue='Gender')
plt.title('2023 Patients blood type count')
plt.legend(bbox_to_anchor =(1.15, 1))
plt.grid();
```

```python
sns.countplot(data=data, x='Medical Condition', hue='Test Results')
plt.title('Medical condition count')
plt.legend(title='Test Result', bbox_to_anchor =(1, 1))
plt.grid();
```

```python
plt.pie(data['Medical Condition'].value_counts(), labels=data['Medical Condition'].unique(), autopct ='%.1f%%')
plt.legend(title='Medical Condition', loc = 4,fontsize  = 12, bbox_to_anchor =(2, 0.42), ncol = 1)
plt.title('Percent of patients Medical condition');
```

```python
sns.histplot(data=data, x='Billing Amount($)')
plt.title('Hospital bill distribution');
```

```python
sns.histplot(data=data, x='Age')
plt.title('Patient age distribution');
```

```python
sns.histplot(data=data, x='Medication')
plt.title('Medication Distribution');
```

```python
plt.figure(figsize=(15,15))

plt.subplot(3,2,1)
sns.countplot(data, x='Test Results')
plt.title('Patients test result count')
plt.ylabel('Number of patients')

plt.subplot(3,2,2)
sns.countplot(data, x='Medication')
plt.title('Patients medication count')
plt.ylabel('Number of patients')

plt.subplot(3,2,3)
sns.countplot(data, x='Admission Type')
plt.title('Patients admisssion type count')
plt.ylabel('Number of patients')

plt.subplot(3,2,4)
sns.countplot(data, x='Medical Condition')
plt.title('Patients medical condition count')
plt.ylabel('Number of patients')

plt.subplot(3,2,5)
sns.countplot(data, x='Insurance Provider')
plt.title('Patients insurance provider count')
plt.ylabel('Number of patients')

plt.subplot(3,2,6)
sns.countplot(data, x='Blood Type')
plt.title('Patients Blood type count')
plt.ylabel('Number of patients')

plt.tight_layout()
plt.show;
```
