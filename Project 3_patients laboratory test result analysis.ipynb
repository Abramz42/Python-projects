# Questions and Answers

## 1. How many female patients at age 18 and above had a normal Haematocrit level?

```python
# F=36-48%  M=40-54% sFM=30-44%

a = data[(data['SEX'] == 'F') & (data['AGE'] >= 18) & (data['HAEMATOCRIT'] >= 36) & (data['HAEMATOCRIT'] <= 48)]
a
```

```python
a.shape
```

```python
print("""
How many female patients at age 18 and above had a normal Haematocrit level?
Answer: {} female patients""".format(a.shape[0]))
```

## 2. How many patients had a normal MCV?

```python
#mean corpuscular volume FM=80-100fl femtolitre
b = data[(data['MCV'] >= 80) & (data['MCV'] <= 100)]
b
```

```python
b.shape
```

```python
print("""
How many patients had a normal MCV?
Answer: {} Patients
""".format(b.shape[0]))
```

## 3. How many in care patients had a abnormal leucocyte level?

```python
#sFM<2=6200-17000/mm FM<2=4.5-11.0cells/l
c = data[(data['SOURCE'] == 'in') & (data['LEUCOCYTE'] >= 4.5) & (data['LEUCOCYTE'] <= 11)]
c
```

```python
c.shape
```

```python
print("""
How many in care patients had a abnormal leucocyte level?
Answer: {} patients""".format(c.shape[0]))
```

## 4. How many male patients had a normal Haematocrit level and MCH?

```python
#27-31picograms/cell
d = data[(data['SEX']=='M')&(data['HAEMATOCRIT']>=40)&(data['HAEMATOCRIT']<=54)&(data['MCH']>=27)&(data['MCH']<=31)]
d
```

```python
d.shape
```

```python
print("""
How many male patients had a normal Haematocrit level and MCH?
Answer: {} Patients""".format(d.shape[0]))
```
